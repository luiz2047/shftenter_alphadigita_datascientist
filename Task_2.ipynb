# Вопрос Д

Для ответа на данный вопрос построим два графика распределения возраста заемщиков для мужчин и женщин

```python
df_man = df[df['X2'] == "1"]
df_woman = df[df['X2'] == "2"]
```

```python
feature = "X5"
ax = df_man[feature].value_counts().sort_index().plot(kind='bar',
                                    figsize=(10,5), 
                                    color='blue',
                                    )
ax.set_xlabel("Возраст")
ax.set_ylabel("Количество")
ax.set_title('Распределение по возрасту(Sex=1(М))')
experiment.log_figure(figure_name="Распределение по возрасту(Sex=1(М))", figure=plt)
```

```python
ax = df_woman[feature].value_counts().sort_index().plot(kind='bar',
                                    figsize=(10,5), 
                                    color='blue',
                                    )
ax.set_xlabel("Возраст")
ax.set_ylabel("Количество")
ax.set_title('Распределение по возрасту(Sex=2(Ж))')
experiment.log_figure(figure_name="Распределение по возрасту(Sex=2(Ж))", figure=plt)
```

Для проверки гипотезы в данном случае можно использовать t-критерий, который подходить для независимых выборок.

```python
a_raspr = df_man[feature].astype(int)
b_raspr = df_woman[feature].astype(int)
```

```python
import scipy.stats
```

```python
scipy.stats.ttest_ind(a_raspr,b_raspr)
```

Получаем результат что pvalue гораздо меньше, чем 0.05. Поэтому нулевая гипотеза верна.
